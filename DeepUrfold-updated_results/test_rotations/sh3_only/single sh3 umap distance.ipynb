```python
import pandas as pd

latent = pd.read_csv("../../superfamily_models_results/2.30.30.100/latent.csv")
```

```python
latent
```

```python
latent = latent.drop(columns=["Unnamed: 0"]).set_index(["index"])
latent
```

```python
import umap
```

```python
reducer = umap.UMAP(n_components=2, metric="cosine")
```

```python
reducer.fit(latent.values)
```

```python
import torch
rotated_latent = torch.load("../1kq2A00_latent-mean-mean.pt")
```

```python
rotated_latent
```

```python
umap_embeddings = reducer.transform(rotated_latent)
```

```python
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
distance_umap = pairwise_distances(umap_embeddings, metric="cosine")
```

```python
distance_umap2 = distance_umap[np.triu_indices(distance_umap.shape[0], 1)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.kdeplot(distance_umap2.flatten())
plt.xlim(-0.08,2.08)
plt.xlabel("Cosine Distance")
#plt.title("Distribution of UMAP 2D Embedding Distances for\n500 Randomly Rotated Copies of 1kq2A00\nSubjected to the SH3 Model")
plt.title("Distribution of Pairwise Distances between\n2D UMAP Embeddings of 500 Randomly Rotated\nCopies of 1kq2A00Subjected to the SH3 Model")
plt.savefig("random_rotation_distanes_1kq2A00_umap_sh3_model.png", dpi=600, bbox_inches="tight")
plt.savefig("random_rotation_distanes_1kq2A00_umap_sh3_model.pdf", dpi=600, bbox_inches="tight")
```
